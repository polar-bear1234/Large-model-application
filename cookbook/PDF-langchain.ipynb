```python
text_splitter = CharacterTextSplitter(
    separator="。", 
    chunk_size=200, 
    chunk_overlap=0, 
    length_function=len
)
docs = text_splitter.split_documents(texts)
print(len(docs))
```

```python
embeddings = OpenAIEmbeddings()
db = FAISS.from_documents(docs, embeddings)
```

```python
query = "What is this document mainly about？"
# docs = db.similarity_search(query)
docs_and_scores = db.similarity_search_with_score(query, k=4)
```

```python
scores = []
contents = []
for d in docs_and_scores:
    scores.append(d[1])
    contents.append(d[0])
sim_ind = np.argmax(scores)
content = contents[sim_ind]
print("最大的相似度为{}.".format(scores[sim_ind]))
print("检索的文本为:", "\n", content.page_content)
```

```python
for score, doc in zip(scores, contents):
    print("相似度-----------", score)
    print(doc.page_content)
    print("\n")
```

```python
from collections import deque

def monotonic_queue(nums, k):
    result = []
    q = deque()
    for i in range(len(nums)):
        while q and nums[i] > nums[q[-1]]:
            q.pop()
        q.append(i)
        while q and q[0] < i - k + 1:
            q.popleft()
        if i >= k - 1:
            result.append(nums[q[0]])
    return result

# 示例
nums = [1, 3, -1, -3, 5, 3, 6, 7]
k = 3
result = monotonic_queue(nums, k)
print(result)
```
